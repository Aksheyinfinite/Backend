```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
initBuffT=pd.read_csv("firetv-INITIAL-BUFFER-TIME.csv")
```

```python
initBuffT.columns
```

```python
initBuffT.DURATION
```

```python
sns.violinplot(initBuffT.DURATION,initBuffT["CONTENT-PARTNER"])
plt.title("FIRE TV Initial Buffer Time")
```

```python
initBuffT.groupby("$region").size().index.values
```

```python
plt.bar(initBuffT.groupby("$region").size().index.values,initBuffT.groupby("$region").size().values)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(initBuffT["CONTENT-PARTNER"])
```

```python
Lcontent=initBuffT.groupby(['$region', 'CONTENT-PARTNER']).size().reset_index().pivot(columns='$region', index='CONTENT-PARTNER', values=0)
Lcontent.plot(kind="bar",stacked=True,figsize=(10,5))
plt.title("Initial Buffer Time")
```

```python
initBuffT.groupby(['$region', 'CONTENT-PARTNER']).size().reset_index().pivot(columns='$region', index='CONTENT-PARTNER',values=0)
```

```python
sns.violinplot(initBuffT.DURATION,initBuffT["$region"])
plt.title("FIRE TV Initial Buffer Time")
```

```python
Lcontent=initBuffT[initBuffT.DURATION>15].groupby(['$region', 'CONTENT-PARTNER']).size().reset_index().pivot(columns='$region', index='CONTENT-PARTNER', values=0)
Lcontent.plot(kind="bar",stacked=True,figsize=(10,5))
plt.title("FIRETV PLAY ONDEMAND FAIL")
```
